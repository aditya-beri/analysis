```python
#pclass: Passenger Class (1 = 1st; 2 = 2nd; 3 = 3rd)
#survived: Survival (0 = No; 1 = Yes)
#name: Name
#sex: Sex
#age: Age
#sibsp: Number of siblings/spouses aboard
#parch: Number of parents/children aboard
#embarked: Port of embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)#

%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
df_train = pd.read_csv(r'C:\Users\ADITYA\Desktop\train.csv')
df_train.head()
```

```python
df_train.groupby('Sex')[['Survived'][SibSp]].mean()
```

```python
q=df_train.groupby('Sex')[['Survived']].mean()
sns.heatmap(q,cmap='magma',linecolor='white',linewidths=1)
#THIS SHOW US THAT AVERAGE SURVIVAL OF WOMEN WAS GREATER THAN THAT OF MEN#
```

```python
df_train.head()
g = sns.FacetGrid(df_train, col="Sex", row="Survived")
```

```python
g = sns.FacetGrid(df_train, col="Sex",  row="Survived")
g = g.map(plt.hist, "SibSp")
#the following plot shows survival of both sex on the basis of SibSp#
#we infer that females with less SibSp had a greater chance of survival
```

```python
sns.stripplot(x="SibSp", y="Sex", data=df_train,jitter=True,hue='Survived',palette='Set2',split=True)
#this plot gives the same information as given by the above plot
```

```python
sns.stripplot(x="Age", y="Sex", data=df_train,jitter=True,hue='Survived',palette='Set1',split=True)
#THIS PLOT SHOWS SURVIVAL OF GENDER BASED UPON THEIR AGES#
```
